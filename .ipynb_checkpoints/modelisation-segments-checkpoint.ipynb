# Création de nouvelles caractéristiques 

```python
#data caracteristics
#create a function that determines the dataset caracteristics
def caracteristics (dataset):
    carac=pd.DataFrame(index=dataset.columns.tolist())
  #Types de variables
    carac['types']=dataset.dtypes.tolist()
  #Nombre de modalites
    modalites=[]
    for i in dataset.columns.tolist():
        modalites.append(len(set(dataset[i].tolist())))
    carac['nombre de modalites']=modalites
  #Nombre de valeurs manquantes
    missing_values=[]
    for i in dataset.columns.tolist():
        missing_values.append(dataset[i].isna().sum())
    carac['number of missing values']=missing_values
  #Pourcentage de valeurs manquantes
    prop_missing_values=[]
    for i in dataset.columns.tolist():
        prop_missing_values.append(round(((dataset[i].isna().sum()/dataset.shape[0])*100),1))
    carac['Proportion number of missing values(%)']=prop_missing_values
  #Moyenne
    moyenne=[]
    for i in dataset.columns.tolist():
        if is_numeric_dtype(dataset[i])==True:
            moyenne.append(dataset[i].mean())
        else:
            moyenne.append (np.nan)
    carac['moyenne']=moyenne
      #Maximum
    maxim=[]
    for i in dataset.columns.tolist():
        if is_numeric_dtype(dataset[i])==True:
            maxim.append(dataset[i].max())
        else:
            maxim.append (np.nan)
    carac['maximum']=maxim
      #Minimum
    minim=[]
    for i in dataset.columns.tolist():
        if is_numeric_dtype(dataset[i])==True:
            minim.append(dataset[i].min())
        else:
            minim.append (np.nan)
    carac['minimum']=minim
#Mode
    mode=[]
    for i in dataset.columns.tolist():
        mode.append(statistics.mode(dataset[i].tolist()))
    carac['mode']=mode
    prop_mode=[]
    for i in dataset.columns.tolist():
        if pd.isna(carac.loc[i,"mode"])==False:
            prop_mode.append(round((dataset[dataset[i]==carac.loc[i,"mode"]][i].value_counts().tolist()[0]/dataset[i].value_counts().sum())*100,1))
        else:
            prop_mode.append(round((dataset[i].isna().sum()/(dataset[i].value_counts().sum()+dataset[i].isna().sum()))*100,1))
    carac['proportion mode (%)']=prop_mode

#Afficher les caractéristiques
    return carac
```

```python
#Créer un fonction pour construire des diagrammes en bar

def add_value_label(x_list, y_list):
    for i in range(1, len(x_list) + 1):
        plt.text(i, y_list[i - 1], y_list[i - 1], ha='left', va='center_baseline')

def barplot(dataset, variable, title):
    df = pd.DataFrame(round((dataset[variable].value_counts() / dataset[variable].value_counts().sum() * 100), 1))
    plt.figure(figsize=(12, 6))
    plot = plt.bar(df.index, df[variable], label=df[variable])
    add_value_label(df.index, df[variable])
    plt.xticks(rotation=90)
    plt.ylim(0, 100)
    plt.title(title)
    plt.show()
    return plot

# Répartition des villes
barplot(data, "ville_x", "Répartition des villes")

# Répartition des segments
barplot(data, "segments_x", "Répartition des segments")
```

```python
#Répartiotn des sexes

def plot_pie_chart(dataset, variable, title):
    df = pd.DataFrame(round((dataset[variable].value_counts() / dataset[variable].value_counts().sum() * 100), 1))
    plt.pie(df[variable], labels=df.index, autopct='%.1f%%')
    plt.title(title)
    plt.show()

# Répartition des sexes
plot_pie_chart(data, "genre_x", "Répartition des sexes")
```

#### Abidjan est la ville la plus représentée et les Collective Engaged le segment le plus représenté. Presque partié des sexes

```python
#Création d'une fonction pour faire des digrammes en barre empilés
def stacked_barplot(data, variable1, variable2):
    tab=pd.crosstab(data[variable1], data[variable2])
    for i in set(list(data[variable1])):
        tab[tab.index==i]=(tab[tab.index==i]/data[data[variable1]==i][variable1].value_counts().values[0])*100
    plot=tab.plot(kind='bar', stacked=True)
    for c in plot.containers:
        # Optional: if the segment is small or 0, customize the labels
        labels = [round(v.get_height(),1) if v.get_height() > 0 else '' for v in c]
        # remove the labels parameter if it's not needed for customized labels
        plot.bar_label(c, labels=labels, label_type='center')
    plt.legend(loc='center')
    return plot
```

# Backward Elimination: Test 5

```python

# Initialiser le modèle d'arbre de décision
dtree = DecisionTreeClassifier(random_state=0)

# Ajuster le modèle sur l'ensemble complet des variables
dtree.fit(train5_x, train5_target)

# Critère d'arrêt (à définir en fonction de vos besoins)
stopping_criterion = 0.02

while np.min(dtree.feature_importances_) < stopping_criterion:
    # Identifier la caractéristique avec la plus faible importance
    min_importance_index = np.argmin(dtree.feature_importances_)
    
    # Supprimer la caractéristique
    train5_x = train5_x.drop(columns=train5_x.columns[min_importance_index])
    
    # Réentraîner l'arbre de décision
    dtree.fit(train5_x, train5_target)
    
    
# Afficher les caractéristiques finales
print(train5_x.columns)
```

```python
# Réentraîner l'arbre de décision
dtree.fit(train5_x, train5_target)
# Prédire sur les ensembles de test
pred_test5_1 = dtree_selected.predict(test5_x1_selected)
pred_test5_2 = dtree_selected.predict(test5_x2_selected)
    
# Évaluer la performance
accuracy_test5_3 = accuracy_score(test5_target1, pred_test5_1)
accuracy_test5_4 = accuracy_score(test5_target2, pred_test5_2)
    
```

# Forward Selection: Test 5

```python
selected_features = []

# Critère d'arrêt (à définir en fonction de vos besoins)
stopping_criterion = 0.01

while np.max(dtree.feature_importances_) > stopping_criterion:
    # Identifier la caractéristique avec la plus grande importance
    max_importance_index = np.argmax(dtree.feature_importances_)
    
    # Ajouter la caractéristique à l'ensemble sélectionné
    selected_features.append(train5_x.columns[max_importance_index])
    
    # Mettre à jour l'ensemble des caractéristiques
    train5_x = train5_x[selected_features]
    
    # Réentraîner l'arbre de décision
    dtree.fit(train5_x, train5_target)

# Afficher les caractéristiques finales
print(selected_features)
```
